###**Joining array**

```python
# vstack() it combines the second array as new rows of the first array.
a =np.ones((3,3))
b = np.zeros((3,3))
```

```python
np.vstack((a,b))
```

```python
# hstack() perform horizontal stacking i.e. the second array is added to the columns of the first array.
np.hstack((a, b))
```

```python
# column_stack() and row_stack() function perform stacking between multiple array.
a = np.array([1,2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])
np.column_stack((a, b, c))
```

```python
np.row_stack((a, b, c))
```

###**Splitting Arrays**

```python
# hsplit() divide an array into several part horizontally.
a = np.arange(16).reshape((4,4))
a
```

```python
[b, c] = np.hsplit(a, 2)
b
```

```python
c
```

```python
# vsplit() divide an array into several part vertically.
[b, c] = np.vsplit(a, 2)
b
```

```python
c
```

```python
# split() split the array into nonsymmeterical parts.
# (axis = 1) for column and (axis = 0) for row.
[a1, a2, a3] = np.split(a,[1,3], axis=1)
a1
```

```python
a2
```

```python
a3
```

```python
[a1, a2, a3] = np.split(a, [1,3], axis=0)
a1
```

##**Reading and Writing Array Data on files**

```python
# Regarding for saving and then later retrieving data stored in binary format, NumPy provides a pair of functions called save() and load().
a = np.array([[1,2,3], [4,5,6]])
np.save('saved_data', a)
```

```python
# load() to load the saved data.
load_data = np.load('saved_data.npy')
load_data
```

**Reading file with Tabular data**

```python
# To be able to read your data in a text file and insert them into an array, NumPy provides a function called genfromtxt().
data = np.genfromtxt('data.csv', delimiter=',', names=True)
```
